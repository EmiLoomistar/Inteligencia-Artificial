```python
class Grafo:
    def __init__(self, adyacencias):
        self.adyacencias = adyacencias
        self.trayectoria_optima = []
    def vecinos(self, v):
        return self.adyacencias[v]

    # Funcion heuristica, aqui se utiliza la distancia a Bucharest
    def distancia_bucharest(self, ciudad):
        distancias_a_bucharest = {
            "Arad": 366,
            "Bucharest": 0,
            "Craiova": 160,
            "Dobreta": 242,
            "Eforie": 161,
            "Fagaras": 178,
            "Giurgiu": 77,
            "Hirsova": 151,
            "Iasi": 226,
            "Lugoj": 244,
            "Mehadia": 241,
            "Neamt": 234,
            "Oradea": 380,
            "Pitesti": 98,
            "Rimnicu Vilcea": 193,
            "Sibiu": 253,
            "Timisoara": 329,
            "Urziceni": 80,
            "Vaslui": 199,
            "Zerind": 374
        }
        return distancias_a_bucharest[ciudad]
    
    def imprime_pasos(self, a_visitar,padres, i, distancia_acumulada=None):
        if distancia_acumulada is not None:
            pasos = {ciudad:distancia_acumulada[ciudad]+self.distancia_bucharest(ciudad) 
                     for ciudad in a_visitar}
            
            pasos_final = {}
            for ciudad, costo in pasos.items():
                trayectoria = []
                #Debido a que la ciudad inicial se referencia a ella misma 
                # mientras el padre no se referencia a el mismo agrega ciudades a trayectoria
                while padres[ciudad] != ciudad: 
                    trayectoria.append(ciudad)
                    ciudad = padres[ciudad]
                #Agregamos la ciudad inicial
                trayectoria.append(ciudad)
                #Invertimos la lista para que se muestre en orden
                trayectoria.reverse()
                #Imprimimos la trayectoria y el costo
                print(f'{i}. {"-".join(trayectoria)} {costo}')

    def algoritmo_a_estrella(self, ciudad_inicio, ciudad_destino):
        # a_visitar es una lista de ciudades proximas que no han sido expandidas
        # visitadas es la lista de ciudades visitadas por lo tanto expandidas
        a_visitar = set([ciudad_inicio])
        visitadas= set([])

        # distancia acumulada es una de las metricas utilizadas para calcular la ruta optima
        # Ej. {"Arad":0, "Sibiu":140...}
        distancia_acumulada = {}
        # Lo iniciamos con la distancia acumulada de la ciudad inicial el cual es 0
        distancia_acumulada[ciudad_inicio] = 0

        # padres es un diccionario utilizado para registar el nodo padre de las ciudades
        # esto es las ciudades de las que venimos ej {"Sibiu": "Arad", "Rimnicu": "Arad"...}
        padres = {}
        padres[ciudad_inicio] = ciudad_inicio
        # Ademas niciamos el diccionario indicando que el padre de la ciudad de inicio es el mismo
        # Esto nos ayudara al momento de generar la trayectoria final
        
        #Variable i para llevar registro del nivel
        i = -1
        ciudad_visitada = ""

        while len(a_visitar) > 0:
            
             #Imprime pasos
            i += 1
            print("Inicio\n" if ciudad_visitada == "" else "", end="")
            self.imprime_pasos(a_visitar,padres, i, distancia_acumulada)
            print("" if ciudad_visitada == "" else f"Ciudad Expandida: {ciudad_visitada}\n")
            
            # Encuentra la ciudad proxima con la menor distancia
            ciudad_actual = None
            for ciudad in a_visitar: #v es el nombre de una ciudad ej 'Arad'
                if ciudad_actual == None or \
                    distancia_acumulada[ciudad] + self.distancia_bucharest(ciudad) \
                    < distancia_acumulada[ciudad_actual] + self.distancia_bucharest(ciudad_actual):
                    ciudad_actual = ciudad;
            if ciudad_actual == None:
                print('No existe trayectoria!')
                return False
            
            # Si la ciudad actual es la destino comienza a generar la trayectoria
            if ciudad_actual == ciudad_destino:
                trayectoria = []
                costo = distancia_acumulada[ciudad_actual]
                #Debido a que la ciudad inicial se referencia a ella misma 
                # mientras el padre no se referencia a el mismo agrega ciudades a trayectoria
                while padres[ciudad_actual] != ciudad_actual: 
                    trayectoria.append(ciudad_actual)
                    ciudad_actual = padres[ciudad_actual]
                #Agregamos la ciudad inicial
                trayectoria.append(ciudad_inicio)
                #Invertimos la lista para que se muestre en orden
                trayectoria.reverse()
                #Imprimimos la trayectoria y el costo
                print(f"\nTrayectoria Optima: {trayectoria}")
                print(f"Costo Total: {costo}")
                self.trayectoria_optima = trayectoria
                return 
            
            # Para todos los vecinos de la ciudad escogida
            for (ciudad, costo) in self.vecinos(ciudad_actual):
                # Si la ciudad no esta en ciudades a visitar o ciudades visitadas
                # agregarla a la ciudad a visitar ,registrar "n" como el padre y el costo
                if ciudad not in a_visitar and ciudad not in visitadas:
                    a_visitar.add(ciudad)
                    padres[ciudad] = ciudad_actual
                    distancia_acumulada[ciudad] = distancia_acumulada[ciudad_actual] + costo
                # De otra forma, checa si es mas rapido visitar n que la ciudad actual
                # Si lo es, actualiza al padre y su distancia acumulada
                else:
                    if distancia_acumulada[ciudad] > distancia_acumulada[ciudad_actual] + costo:
                        distancia_acumulada[ciudad] = distancia_acumulada[ciudad_actual] + costo
                        padres[ciudad] = ciudad_actual
                        # Y si el nodo estaba en ciudades visitadas, lo mueve a ciudades a visitar
                        if ciudad in visitadas:
                            visitadas.remove(ciudad)
                            a_visitar.add(ciudad)
                            
            # Remueve la ciudad actual de las ciudades a visitar y lo agrega a visitadas
            a_visitar.remove(ciudad_actual)
            visitadas.add(ciudad_actual)
            ciudad_visitada = ciudad_actual
            
        print('No existe una ruta valida')
        return
```
